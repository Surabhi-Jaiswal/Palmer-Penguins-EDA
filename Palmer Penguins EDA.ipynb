```python
df['bill_length_mm'].fillna(df['bill_length_mm'].mean(), inplace=True)
df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean(), inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean(), inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].mean(), inplace=True)
```

```python
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df, palette='pastel')
plt.title('Species Distribution')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='island', data=df, palette='coolwarm')
plt.title('Island Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['bill_length_mm'], kde=True, bins=20, color='teal')
plt.title('Bill Length Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['bill_depth_mm'], kde=True, bins=20, color='orange')
plt.title('Bill Depth Distribution')

plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['flipper_length_mm'], kde=True, bins=20, color='purple')
plt.title('Flipper Length Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['body_mass_g'], kde=True, bins=20, color='green')
plt.title('Body Mass Distribution')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df, palette='Set2')
plt.title('Bill Length vs Bill Depth by Species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=df, palette='Set1')
plt.title('Flipper Length vs Body Mass by Species')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', data=df, palette='mako')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.pairplot(df, hue='species', diag_kind='kde', palette='husl')
plt.suptitle('Pairplot of Palmer Penguins Data', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='year', data=df, palette='cividis')
plt.title('Penguin Observations over the Years')
plt.show()
```
